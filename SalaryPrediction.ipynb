# Clean Up DATA:

### Remove GrossPay (Very Similar to AnnualSalary)

```python
df.drop('GrossPay', axis=1, inplace=True)
```

#### Looking at the creepy ends of data, I could see that two fields need to be modified:

- HireDate is separated by ':', '-' and '/'. So I can convert them all to '/' to keep it uniform.

- Annual Salary needs to be converted to a useful form, i.e., float data type.

- Rest of the columns seem to be alright. Only 'name'  column I will replace with 'Gender' and 'AgencyID' column will also be dropped because it is very similar to 'Agency' column. 


## Clean HireDate Column

```python
df['HireDate'] = df['HireDate'].astype('str') 
```

```python
(df['HireDate'])[:5]
```

```python
df['HireDate'] = df['HireDate'].apply(lambda date: date.replace('-', '/') if '-' in date else date.replace(':', '/'))
```

## Start the DECISION TREE Part:

### I think I should get a reasonably good prediction about the salary slab because:

- Firstly, it is a slab, not continuous values of data
- The columns such as JobTitle, Agency and HireDate feel like pretty sturdy measures for finding the paySlab and I expect the salaries of a particular deparment should not vary too much.
- But, I'm not sure! Anything can happen.

Remove all duplicate records

## Convert column data into numbers and APPEND to Parameters X

```python
X = pd.get_dummies(df2['JobTitle'], prefix = 'JobTitle')
```

```python
X = pd.concat([X, pd.get_dummies(df2['Agency'], prefix = 'Agency')], axis=1)
```

```python
X = pd.concat([X, pd.get_dummies(df2['HireDate'], prefix = 'HireDate')], axis=1)
```

```python
X = pd.concat([X, pd.get_dummies(df2['Gender'], prefix = 'Gender')], axis=1)
```

```python
X.head(5)
```

## Now let's get a Y dataframe based on the Pay Slab (salary range).

### We take the PaySlab as the Y because the aim is to predict the salary range. The pay slab is basically the salary range, and finding the payslab helps in Desicion Trees for the following reasons:

- Having an extremely large number of classes will make the treeenormous because the classes will be real valued.

- Having the exact salary predicted is not so useful if you are trying to match a number, it is more intuitive to find a range for salary because it maps a profession and will let you know earning potential.


- I did not choose other fields for the Y because in this dataset, either we can predict salary from the profession or profession from salary. But since we have been given Agency,  AgencyID and so on, it makes more sense to use this for salary as it would be sort of like cheating to guess the profession while knowing AgencyID and vice versa. Also, the coulumn HireDate would not have any specific use for guessing profession, the way it does in guessing salary.


```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
Y = le.fit_transform(df2['PaySlab'])
```

```python
le.inverse_transform([0, 1, 2, 3, 4, 5 ,6 ,7, 8])
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X.values, Y)
```

```python
X.columns
```

```python
with open("dt1.dot", 'w') as f:
    export_graphviz(dt, out_file=f, feature_names=X.columns)
```

### Image is too large to be displayed, so limiting the depth to 4

```python
dt = DecisionTreeClassifier(max_depth = 4)
dt.fit(X.values, Y)
with open("dt2.dot", 'w') as f:
    export_graphviz(dt, out_file=f, feature_names=X.columns)
os.system('dot -Tpng dt2.dot -o dt2.png')
```

```python
Image(filename='dt2.png')
```

I feel the Tree has learned the best way to split the data is using Profession or 'JobTitle' -> Department -> HireDate

### I feel this tree makes some sense because:
- Firstly, I can see few pure classes formed at the depth of 2 and 3 itself in the tree, which means something is working. 
- At the root, the two partitions are about 12000 and 2000, which though varying quite a bit seems to be a good beginning considering the size of the data and variability of the fields
- Even at depth 4, if you leave out the left most node, the remaining have quite some pure classes created
- Most of the Decisions are taking according to the Profession and Agency, which is what I had expected
- The logic of the tree seems to be: first take profession, next check on department and finally use the hiring date to make the final pure class. This logic seems sound and I think if the depth of the tree is increases, we will get a good accuracy for this dataset as well

## Do CROSSVALIDATION to test quality of Results

```python
from sklearn import cross_validation
```

```python
scores = cross_validation.cross_val_score(dt, X, Y, cv = 10)
```

```python
scores
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

# Improve accuracy by increasing the depth of the Decision Tree

```python
dt = DecisionTreeClassifier(min_samples_split = 100)
scores = cross_validation.cross_val_score(dt, X, Y, cv = 10)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

The accuracy of the tree has increased because of increasing the depth of the tree as:
- The learning of the tree is better due to more decision nodes
- More fields are taken into consideration as the depth of the tree increases
- Purer data samples are formed at greater depths
